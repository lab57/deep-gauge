```python
import tensorflow as tf

from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
```

```python
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
train_labels
```

```python
import os
import numpy as np
from PIL import Image


def get_images_and_labels(root_folder):
    images = []
    labels = []
    # iterate through all directories inside the root directory
    for directory in os.listdir(root_folder):
        print(directory)
        if os.path.isdir(os.path.join(root_folder, directory)):
            # parse the directory name to get the label
            label = float(directory.split("_")[-1].replace("-", "."))

            # iterate through all the files inside the directory
            for filename in os.listdir(os.path.join(root_folder, directory)):
                # ensure we're working with an image file
                if filename.endswith(".png") or filename.endswith(".jpg"):
                    # open the image file and convert it to a numpy array
                    if(filename == "gauge_0.png"):
                        continue
                    img = Image.open(os.path.join(root_folder, directory, filename))
                    img_array = np.array(img)

                    # append the image array and label to their respective lists
                    images.append(img_array)
                    labels.append(label)

    return images, labels


root_folder = "./images/tf_gauges/"
images, labels = get_images_and_labels(root_folder)

# Now 'images' is a list of numpy arrays and 'labels' is a list of corresponding labels
labels
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(10))
model.summary()
```

plt.imshow()

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(train_images, train_labels, epochs=10, 
                    validation_data=(test_images, test_labels))
```
